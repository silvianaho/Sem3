9. Use the gradient descent algorithm to find the minimum point of $f(x, y) = \cos(0.125x)^{3} - \sin(y)$ with the following settings:

Set epsilon = 0.0001 for all three cases and 500 as maximum number of iterations. Use the
difference in values of $f(x,y)$ less than epsilon as the convergence criterion. Compare and
comment on your results. 


```python
import numpy as np
import sympy as sp

print("numpy version {}".format(np.__version__))
print("simpy version {}".format(sp.__version__))
```

```python
epsilon = 0.0001
max_iter = 500
f = lambda x, y: np.cos(0.125*x**3) - np.sin(y)
partial_x = lambda x, y: -0.375 * x**2 * np.sin(0.125*x**3)
partial_y = lambda x, y: -np.cos(y)
```

```python
x=sp.Symbol('x') 
y=sp.Symbol('y')
sp.diff(sp.cos(0.125*x**3) - sp.sin(y), x)
```

(a) Initial point: $x=2.5, y=1$, learning rate $\alpha=0.1$

```python
x = 2.5
y = 1
alpha = 0.1

for n in range(max_iter):
    x_n = x - alpha * partial_x(x, y)
    y_n = y - alpha * partial_y(x, y)
    print("Iteration", n+1, ": x = ", x_n, ", y = ", y_n, ", f(x,y) = ", f(x_n, y_n))
    diff = abs(f(x_n, y_n)-f(x, y))
    x = x_n
    y = y_n
    if diff < epsilon:
        print("The local minimum occurs at iteration {} x={}, y={}, f(x,y)={}".format(n+1, x, y, f(x,y)))
        break
    if n+1 == max_iter:
        print("Did not converge after {} iterations".format(max_iter))
```

(b) Initial point: $x=-2, y=-4$, learning rate $\alpha=0.1$
