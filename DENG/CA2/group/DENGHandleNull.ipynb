```python
import numpy as np
import pandas as pd
```

```python
headers = ['customerNumber', 'customerName', 'contactLastName', 'contactFirstName', 
           'phone', 'addressLine1', 'addressLine2', 'city', 'state', 'postalCode', 
           'country', 'salesRepEmployeeNumber', 'creditLimit']
data = pd.read_csv('./Customers.csv', names=headers, encoding='iso-8859-1')
```

```python
# data[data.isna().any(axis=1)] # get all row with null values
# data[pd.isnull(data.salesRepEmployeeNumber)] # get all row with null salesRepEmployeeNumber
```

```python
# data = data.fillna({'salesRepEmployeeNumber': ''})
data = data.fillna({'salesRepEmployeeNumber': 0})
```

```python
data[pd.isnull(data.salesRepEmployeeNumber)]
```

```python
data.head(10)
```

```python
data.salesRepEmployeeNumber = pd.to_numeric(data.salesRepEmployeeNumber, errors='coerce', downcast='signed')
```

```python
data.salesRepEmployeeNumber = data.salesRepEmployeeNumber.astype('int', errors='ignore')
```

```python
data.salesRepEmployeeNumber = data.salesRepEmployeeNumber.replace(0, '')
```
